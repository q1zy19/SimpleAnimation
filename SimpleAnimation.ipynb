```python
import tkinter as tk
import random

class Raindrop:
    def __init__(self, canvas):
        self.canvas = canvas
        self.id = canvas.create_oval(2, 10, 0, 0, fill=random.choice(["aqua","aliceblue","blue"]),outline='lightgrey')
        self.x = random.randint(0, int(canvas['width']))
        self.y = random.randint(0, int(canvas['height'])) - 100
        self.speed = random.randint(5, 15)
        self.size = random.randint(5, 20)
        self.canvas.move(self.id, self.x, self.y)
    
    def move(self):
        self.y += self.speed
        self.canvas.move(self.id, 0, self.speed)
        if self.y > int(self.canvas['height']):
            self.canvas.delete(self.id)
            return True
        return False

class App:
    def __init__(self, master):
        self.master = master
        self.canvas = tk.Canvas(master, width=500, height=500, bg="lightgrey")
        self.canvas.pack()
        self.raindrops = []
        self.create_raindrops()
        self.master.after(10, self.update)
        
    def create_raindrops(self):
        for i in range(10):
            self.raindrops.append(Raindrop(self.canvas))
            
    def update(self):
        for raindrop in self.raindrops:
            if raindrop.move():
                self.raindrops.remove(raindrop)
                del raindrop
        self.create_raindrops()
        self.raindrops = sorted(self.raindrops, key=lambda x: x.y)
        self.master.after(10, self.update)

root = tk.Tk()
app = App(root)
root.mainloop()
```

```python
import tkinter as tk
import math

class Planet:
    def __init__(self, canvas, x, y, radius, distance, speed, color):
        self.canvas = canvas
        self.distance = distance
        self.speed = speed
        self.color = color
        self.radius = radius
        self.angle = 0
        self.orbit = canvas.create_oval(0, 0, distance*2, distance*2, outline='grey')
        self.id = canvas.create_oval(0, 0, radius*2, radius*2, fill=color)
        self.move(x, y)

    def move(self, x, y):
        self.x = x + math.cos(self.angle) * self.distance
        self.y = y + math.sin(self.angle) * self.distance
        self.canvas.coords(self.id, self.x - self.radius, self.y - self.radius, self.x + self.radius, self.y + self.radius)
        self.canvas.coords(self.orbit, x - self.distance, y - self.distance, x + self.distance, y + self.distance)
        self.angle += self.speed

class SolarSystem:
    def __init__(self, master):
        self.master = master
        self.canvas = tk.Canvas(master, width=600, height=600)
        self.canvas.pack()
        self.planets = []
        self.sun = self.canvas.create_oval(290, 290, 310, 310, fill="yellow")
        self.planets.append(Planet(self.canvas, 300, 300, 10, 100, 0.03, "blue"))
        self.planets.append(Planet(self.canvas, 300, 300, 15, 150, 0.02, "green"))
        self.planets.append(Planet(self.canvas, 300, 300, 20, 200, 0.01, "red"))

    def animate(self):
        for planet in self.planets:
            planet.move(300, 300)
        self.master.after(50, self.animate)

root = tk.Tk()
root.title("Solar System Simulation")
solar_system = SolarSystem(root)
solar_system.animate()
root.mainloop()
```
